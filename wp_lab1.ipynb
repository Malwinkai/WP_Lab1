## 2. Typy danych
Komputer przechowuje wszystkie dane (liczby, teksty, instrukcje programu itp.) w swojej pamięci operacyjnej (RAM) pod postacią ciągów bitów.
Interpretacja danych: To, co oznacza ciąg bitów, zależy od typu danych i interpretacji przez program. Na przykład bajt 01000001 może oznaczać liczbę 65 w kodzie dziesiętnym, albo znak 'A' w kodzie ASCII – decyduje o tym kontekst. Dlatego w językach programowania wprowadzono pojęcie typów danych, aby określić, jak traktować dane bitowe w pamięci.
W języku C podstawowe typy to m.in.:
- **int** – liczba całkowita (integer). Zajmuje zazwyczaj 4 bajty (32 bity), co dla typu signed int daje zakres około od –2 147 483 648 do 2 147 483 647. Istnieją także modyfikatory: short (zwykle 2 bajty), long (8 bajtów) – dające odpowiednio mniejsze lub większe zakresy.
- **char** – typ znakowy (character). Zwykle przechowuje kod znaku ASCII. W praktyce używamy char głównie do przechowywania pojedynczych znaków.
- **float** – liczba zmiennoprzecinkowa pojedynczej precyzji, zajmuje 4 bajty, dokładność 6-7 cyfr.
- **double** – liczba zmiennoprzecinkowa podwójnej precyzji. 8 bajtów, dokładność ok. 15–16 cyfr znaczących.
- **bool** – typ logiczny (ang. boolean) reprezentujący wartość prawda/fałsz. Zmienna logiczna przyjmuje wartość true (prawda) lub false (fałsz). W rzeczywistości przechowywana jest zazwyczaj w 1 bicie (gdzie 0 oznacza false, a 1 oznacza true).

## 3. Zmienne

Zmienne to podstawowe pojęcie w programowaniu – oznaczają nazwane miejsca w pamięci, w których możemy przechować dane określonego typu. Każda zmienna ma:
- typ,
 - nazwę nadaną przez programistę (powinny zaczynać się literą/underscore, mogą zawierać cyfry, nie używamy spacji ani polskich znaków). Dobrze, by nazwy były czytelne i opisowe – np. suma, indeks, pozycjaX zamiast nic nie mówiących a, b, c.
- wartość (która może się zmieniać podczas działania programu).

**Deklaracja i inicjalizacja**: W języku C zanim użyjemy zmiennej, musimy ją zadeklarować, podając typ i nazwę, ewentualnie z nadaniem początkowej wartości (inicjalizacją).

### 3.1 Zmienne lokalne i globalne

**Zmienne globalne** (deklarowane poza funkcjami) – te istnieją przez cały czas działania programu.

**Zmienne lokalne** - tworzone wewnątrz funkcji, istnieją tylko podczas wykonywania danej funkcji (po jej zakończeniu pamięć może być użyta ponownie do innych celów). Ich zakres ważności ograniczony jest do bloku { ... }, w którym zostały zadeklarowane.

## Zad 2.

```python
#include <stdio.h>

// Zmienna globalna – istnieje przez cały czas działania programu
int licznik = 0;

void funkcja() {
    // Zmienna lokalna – tworzona na nowo przy każdym wywołaniu funkcji
    int licznik_funkcji = 5;
    printf("W funkcji: licznik_funkcji = %d\n", licznik_funkcji);
}

int main() {
    printf("W main (globalna): licznik = %d\n", licznik);
    funkcja();
    printf("Po powrocie do main: licznik = %d\n", licznik);
    return 0;
}
```

## 4. Operatory

**Operatory** to symbole oznaczające wykonywanie pewnych operacji na danych:
- **Operatory przypisania**: = służy do przypisania wartości do zmiennej. Po lewej stronie musi być zmienna, po prawej wyrażenie. Np. x = 5, y = a + 3. Istnieją też skrócone formy jak +=, -=, *=, /=, %= które łączą operację z przypisaniem (np. x += 5; oznacza x = x + 5;).
- **Operatory porównania** (relacyjne): służą do porównywania wartości, zwracają wynik typu logicznego true/false. Mamy:== (równość porównania, uwaga: dwa znaki = odróżniają od pojedynczego = przypisania),!= (różne),< (mniejsze), > (większe), <= (mniejsze lub równe), >= (większe lub równe).

- **Operatory arytmetyczne**: służą do wykonywania podstawowych działań matematycznych:+ (dodawanie), - (odejmowanie), * (mnożenie), / (dzielenie), % (modulo – reszta z dzielenia).Przykłady: z = x + y;, a = b * 2;, c = c - 1;.

📌Uwaga: operator % jest stosowalny tylko do liczb całkowitych (np. 7 % 3 to 1).

📌 Dzielenie / dwóch liczb całkowitych zwraca część całkowitą wyniku (np. 7 / 3 to 2 w C). Jeśli chcemy uzyskać wynik rzeczywisty, przynajmniej jeden operand musi być typu zmiennoprzecinkowego (np. 7.0 / 3 to  2.3333).
- **Operatory inkrementacji/dekrementacji**: ++ zwiększa wartość zmiennej o 1, a -- zmniejsza o 1. Można ich używać w formie prefiksowej (++i, --i) lub postfiksowej (i++, i--), które różnią się momentem zwrócenia wartości.
